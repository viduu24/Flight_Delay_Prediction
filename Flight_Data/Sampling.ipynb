```python
import pandas as pd
import numpy as np
```

```python
flight_data = pd.read_csv('./flight_data_2024.csv')
flight_data.head()
```

```python
flight_data.info()
```

```python
flight_data.isna().sum()
```

```python
flight_data.drop(columns=['cancellation_code'], inplace=True)
# REmove 'taxi_out', 'wheels_off',  'wheels_on', 'taxi_in'
flight_data.drop(columns=['taxi_out', 'wheels_off',  'wheels_on', 'taxi_in'], inplace=True)
# Remove origin, dest
flight_data.drop(columns=['origin', 'dest'], inplace=True)
# Remove diverted and distance column
flight_data.drop(columns=['diverted', 'distance'], inplace=True)
```

```python
flight_data.dropna(inplace=True)
```

```python
flight_data.isna().sum()
```

```python
flight_data.info()
```

```python
sampled_df = (
    flight_data
    .groupby('month', group_keys=False)
    .apply(lambda x: x.sample(frac=0.2, random_state=42))
)

sampled_df = sampled_df.reset_index(drop=True)
```

```python
sampled_df.info()
```

```python
sampled_df.head()
```

```python
flight_data['month'].value_counts(normalize=True)
```

```python
sampled_df['month'].value_counts(normalize=True)
```

```python
# Save as csv flight_data_sample.csv
sampled_df.to_csv('./flight_data_sample.csv', index=False)
```
